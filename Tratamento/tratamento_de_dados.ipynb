## <font color="008080">**Introdução**</font>

O **tratamento dos dados** em modelos de aprendizado de máquina é fundamental. O tratamento inclui manipulações realizadas nos dados, como limpeza, transformação e organização, a fim de aumentar a precisão e a confiabilidade do modelo.

## <font color="008080">**Tratamento dos Dados**</font>

O primeiro passo para realizar o tratamento dos dados que serão utilizados é importar as bilbiotecas que serão utilizadas. Nesse caso, utilizaremos apenas uma biblioteca.

- **Pandas**: O pandas fornece rapidamente estruturas de dados flexíveis e expressivas, projetadas para tornar o trabalho com dados mais fácil e intuitivo.


Em seguida, é necessário carregar o arquivo que contém os dados. 

```python
import pandas as pd

df = pd.read_csv("Students Social Media Addiction.csv")
```

A primeira análise importante a se realizar em um conjunto de dados é a presença de valores faltantes.

Uma das formas de fazer isso é usando `isnull()` e `sum()` combinados.

```python
# Verificar valores nulos
print(df.isnull().sum())
```

Nesse caso, como não temos nenhum valor faltante, não é necessário realizar nenhuma manipulação. Mas, em geral, utilizamos a função `dropna` para excluir linhas que contenham esses valores.

Outras manipulações importantes para os conjuntos de dados são realizadas pelas funções `drop_duplicates` e `convert_dtypes`. A primeira verifica se existem dados duplicados e exclui um deles, a fim de evitar o vazamento desses dados. Já a segunda, converte as colunas em tipos de dados mais apropriados para as predições.

Para realizar o tratamento dos dados, é importante criar uma cópia do Data Frame, para que os dados originais não sofram nenhuma alteração e possam ser recuperados caso necessário.

```python
df_copia = df   # criando a cópia do df

df_sem_dados_faltantes = df_copia.dropna(axis=0)   # excluindo dados faltantes
df_copia = df_sem_dados_faltantes   # atualizando o df

df_sem_dados_duplicados = df_copia.drop_duplicates()   # excluindo dados duplicados
df_copia = df_sem_dados_duplicados   # atualizando o df

df_dados_convertidos = df_copia.convert_dtypes()   #convertendo os tipos dos dados
df_copia = df_dados_convertidos   # atualizando o df
```

Depois de tratar os dados, deve-se salvar as alterações em um novo arquivo, para que possa ser usado novamente em outros códigos.

```python
df_copia.to_csv('dados_tratados.csv', index=False)
```
